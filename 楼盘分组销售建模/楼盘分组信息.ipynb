```python
Dict={}
for i in range(1,41):
    key=df.loc[i].loc['综合评分']
    name=df.loc[i].loc['楼盘']
    stocking=df.loc[i].loc['库存']
    Dict[key]=[name,stocking]
```

```python
print(Dict)
len(Dict)
```

```python
Dict1=OrderedDict(sorted(Dict.items(),key=lambda t:t[0],reverse=True))
Dict1
```

```python
keys=[]
values=[]
for k,v in Dict1.items():
    keys.append(k)
    values.append(v)
```

```python
keys[0]
```

```python
for i in range(0,11):
    score=keys[i]+keys[19-i]+keys[20+i]+keys[39-i]
    group=values[i][0]+'/'+values[19-i][0]+'/'+values[20+i][0]+'/'+values[39-i][0]
    stocking=values[i][1]+values[19-i][1]+values[20+i][1]+values[39-i][1]
    print('分组综合得分%s，组内楼盘:%s,楼盘总库存:%s'%(score,group,stocking))
```
